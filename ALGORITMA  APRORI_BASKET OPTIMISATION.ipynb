Pada kode ini, kita membuat tabel pivot dengan menjalankan fungsi pd.pivot_table() dengan parameter sebagai berikut:
1.	data: DataFrame yang ingin dibuat tabel pivotnya, yaitu df1.
2.	index: Kolom yang akan dijadikan sebagai indeks tabel pivot, yaitu kolom 'InvoiceNo' pada df1.
3.	columns: Kolom yang akan dijadikan sebagai kolom-kolom tabel pivot, yaitu kolom 'Description' pada df1.
4.	values: Kolom yang akan diambil nilainya untuk dihitung pada tabel pivot, yaitu kolom 'Quantity' pada df1.
5.	aggfunc: Fungsi agregasi yang digunakan untuk menghitung nilai tabel pivot, yaitu fungsi sum yang akan menjumlahkan nilai 'Quantity' untuk setiap kombinasi 'InvoiceNo' dan 'Description'.
6.	fill_value: Nilai pengganti yang akan digunakan jika terdapat nilai kosong pada tabel pivot, yaitu 0.
Jadi, kode tersebut akan menghasilkan tabel pivot basket yang berisi nilai jumlah 'Quantity' untuk setiap kombinasi 'InvoiceNo' dan 'Description' pada DataFrame df1.


```python
#mengonversi data dalam format yang diperlukan
#mengonversi menggunakan tabel pivot dan Jumlah jumlah sebagai nilai. isi 0 jika ada nilai nan

basket = pd.pivot_table(data=df1,index='InvoiceNo',columns='Description',values='Quantity', aggfunc='sum',fill_value=0)
```

Fungsi Python yang disebut convert_into_binary yang mengambil satu parameter masukan x.

Fungsi memeriksa apakah nilai input x lebih besar dari nol. Jika ya, fungsi akan mengembalikan nilai 1. Jika tidak, jika nilai inputnya nol atau negatif, fungsi akan mengembalikan nilai 0.

Fungsi ini pada dasarnya mengubah nilai input positif apa pun menjadi 1, dan nilai non-positif apa pun (yaitu nol atau negatif) menjadi 0.

```python
# kita tidak membutuhkan jumlah kuantitas 
# kita perlu mengambil atau tidak
# jadi jika pengguna telah mengambil tanda item itu sebagai 1 tanda lain sebagai 0.

def convert_into_binary(x):
    if x > 0:
        return 1
    else:
        return 0
```

Baris kode ini digunakan untuk menerapkan fungsi convert_into_binary ke setiap elemen dalam DataFrame panda yang disebut basket, dan menyimpan nilai biner yang dihasilkan dalam DataFrame baru yang disebut basket_sets.

Metode .applymap() digunakan untuk menerapkan fungsi convert_into_binary ke setiap elemen dalam keranjang DataFrame. Nilai biner yang dihasilkan kemudian disimpan dalam DataFrame baru yang disebut basket_sets.

Dengan asumsi bahwa fungsi convert_into_binary berfungsi seperti yang dijelaskan dalam jawaban sebelumnya, kode ini akan mengonversi setiap elemen dalam keranjang DataFrame menjadi nilai biner 0 atau 1, bergantung pada apakah nilai aslinya nol atau positif.

Kode ini dapat berguna untuk mengonversi nilai numerik dalam DataFrame menjadi nilai biner, yang dapat berguna untuk jenis tugas analisis atau pembelajaran mesin tertentu.

```python
basket_sets = basket.applymap(convert_into_binary)
```

Kode frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True) adalah implementasi dari algoritma Apriori untuk menghasilkan frequent itemsets dari data.
1.	Algoritma Apriori adalah algoritma populer untuk analisis asosiasi yang menghasilkan himpunan itemset-itemset yang sering muncul dalam data.
2.	Parameter pertama basket_sets adalah input data untuk algoritma Apriori, yang dapat berupa array, DataFrame, atau struktur data lainnya.
3.	Parameter kedua min_support menentukan support minimum yang diperlukan untuk dikategorikan sebagai frequent itemset. Support mengukur seberapa sering suatu itemset muncul dalam data, dan min_support menentukan batas bawah support yang dianggap cukup tinggi untuk dianggap frequent.
4.	Parameter ketiga use_colnames diatur menjadi True untuk menentukan bahwa nama kolom digunakan dalam representasi itemset yang dihasilkan, bukan indeks itemset.
5.	Hasil frequent itemsets kemudian disimpan dalam variabel frequent_itemsets, yang biasanya berupa DataFrame dengan dua kolom: "support" dan "itemsets".
6.	Kolom "support" menunjukkan support dari itemset yang dihasilkan, sedangkan kolom "itemsets" berisi daftar itemset yang dianggap frequent dalam data.


```python
# panggil fungsi apriori dan berikan dukungan minimum di sini kita melewati 7%.
# berarti 7 kali jumlah total transaksi item harus ada.
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)
```

Kode frequent_itemsets merupakan sebuah variabel yang kemungkinan besar merupakan hasil dari suatu analisis data menggunakan algoritma association rule mining, seperti Apriori atau FP-Growth. Variabel ini berisi himpunan itemset-itemset yang muncul dalam data dengan frekuensi tertentu, yang dianggap sebagai "frequent" atau sering muncul dalam data. Itemset-itemset ini dapat terdiri dari satu item saja (singelton) atau beberapa item (multi-item), dan sering kali disimpan dalam bentuk struktur data seperti DataFrame atau list. Jadi, variabel frequent_itemsets ini mungkin berisi daftar itemset-itemset yang sering muncul dalam data dan kemungkinan digunakan untuk menghasilkan aturan asosiasi.

```python
#menghasilkan kumpulan item yang sering
frequent_itemsets
```

1.	rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
	Baris kode ini memanggil fungsi association_rules() untuk melakukan analisis asosiasi pada itemset yang ditemukan sebelumnya dalam variabel frequent_itemsets.
	Fungsi association_rules() adalah fungsi dari library mlxtend yang digunakan untuk menghasilkan aturan asosiasi dari itemset yang diberikan.
	Pada baris ini, parameter metric diatur menjadi "lift", yang menunjukkan bahwa perhitungan kinerja aturan asosiasi didasarkan pada penggunaan ukuran keterkaitan "lift".
	Parameter min_threshold diatur menjadi 1, yang menunjukkan bahwa hanya aturan asosiasi dengan nilai "lift" lebih besar dari 1 saja yang akan disertakan dalam output.
	Output dari fungsi ini kemudian disimpan dalam variabel "rules_mlxtend".
2.	rules_mlxtend.head()
	Baris kode ini menggunakan fungsi head() untuk menampilkan lima baris pertama dari DataFrame yang disimpan dalam variabel "rules_mlxtend".
	Output ini menunjukkan beberapa atribut untuk aturan asosiasi, termasuk itemset asal, itemset tujuan, nilai support, nilai confidence, dan nilai lift.


```python
# Kami akan menerapkan aturan asosiasi pada frequent itemset.
# di sini kita menyetel berdasarkan lift dan mempertahankan lift minimum sebagai 1

rules_mlxtend = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules_mlxtend.head()
```
